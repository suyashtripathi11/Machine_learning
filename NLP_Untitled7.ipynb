```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as ans
```

```python
df = pd.read_csv('train.txt',sep =';',header = None,names=['text','emotion'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
unique_emotion = df['emotion'].unique()
emotion_number = {}
i =0
for emo in unique_emotion:
    emotion_number[emo] = i
    i+=1
df['emotion'] = df['emotion'].map(emotion_number)
```

```python
df
```

```python
df['text'] = df['text'].apply(lambda x: x.lower())
```

```python
import string
def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df['text'] = df['text'].apply(remove_punc)
```

```python
def remove_number(txt):
  new =""
  for i in txt:
    if not i.isdigit():
      new = new+i
    return new
df['text'] = df['text'].apply(remove_number)
```

```python
def remove_emojie(txt):
  new = ""
  for i in txt:
    if i.isascii():
      new+=i
    return new
df['text'] = df['text'].apply(remove_emojie)

```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python

nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words= set(stopwords.words('english'))
```

```python
df.loc[1]['text']
```

```python
def remove(txt):
  words = txt.split()
  cleaned = []
  for i in words:
    if i in stop_words:
      cleaned.append(i)
  return ' '.join(cleaned)
```

```python
df['text'] = df['text'].apply(remove)
```
