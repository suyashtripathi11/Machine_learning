```python
df_encode = pd.get_dummies(df,drop_first=True)
```

```python
df_encode = df_encode.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
numerical_cols = ['Age','RestingBP','Cholestrol_mean','MaxHR']
scaler = StandardScaler()
df_encode[numerical_cols] = scaler.fit_transform(df_encode[numerical_cols])
df_encode.head()
```

```python
df_encode = df_encode.astype(int)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
x = df_encode.drop('HeartDisease',axis =1)
y = df_encode['HeartDisease']
```

```python
x_trian,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
scaler = StandardScaler()
x_trian = scaler.fit_transform(x_trian)
x_test = scaler.transform(x_test)
```

```python
models = {
    "LogisticRegression" : LogisticRegression(),
    "KNeighborsClassifier" : KNeighborsClassifier(),
    "SVC" : SVC(),
    "DecisionTreeClassifier" : DecisionTreeClassifier(),
    "Naive Byeas": GaussianNB(),
    "RandomForestClassifier" : RandomForestClassifier()
}
```

```python
result=[]
```

```python
from sklearn.metrics import f1_score # Added to resolve NameError
for name,model in models.items():
  model.fit(x_trian,y_train)
  y_pred = model.predict(x_test) # Changed x_test_scaled to x_test
  accuracy = accuracy_score(y_test,y_pred) # Changed y to y_pred
  f1 = f1_score(y_test,y_pred)
  result.append({
      'model_name':name,
      'accuracy':round(accuracy,4), # Changed acc to accuracy
      'f1_score':round(f1,4)
  })
```

```python
result
```

```python
import joblib
joblib.dump(models['KNeighborsClassifier'],'KNN_heart.pkl') # Corrected key from 'KNN' to 'KNeighborsClassifier'
joblib.dump(scaler,'scaler.pkl')
joblib.dump(x.columns.tolist(),'columns.pkl')
```
