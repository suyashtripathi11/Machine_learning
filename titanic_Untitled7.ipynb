```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split

# --- Data Preprocessing to ensure x_trian is clean for StandardScaler ---
# Ensure age column has no NaNs
df["age"].fillna(df["age"].mean(), inplace=True)

# Ensure embarked column has no NaNs (fill with mode as it's categorical)
df["embarked"].fillna(df["embarked"].mode()[0], inplace=True)

# Encode categorical columns 'sex' and 'embarked'
le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'])
df["embarked"] = le.fit_transform(df["embarked"])

# Re-define x and y from the now fully preprocessed df
x = df.drop("survived", axis=1)
y = df["survived"]

# Re-do the train-test split with the updated x and y
x_trian, x_test, y_trian, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# --- End Data Preprocessing ---

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_trian)
x_test_scaled = scaler.transform(x_test)

model_svm.fit(x_train_scaled,y_trian)
```

```python
y_pred_svm = model_svm.predict(x_test_scaled)
```

```python

accuracy_score(y_test,y_pred_svm)
```

```python
confusion_matrix(y_test,y_pred_svm)
```

```python
print(classification_report(y_test,y_pred_svm))
```
